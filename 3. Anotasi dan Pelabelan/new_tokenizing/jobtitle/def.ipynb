```python
import mysql.connector  
import time

counter_row = 40
```

```python
# ----------------------------------------------------------   
counter_row -= 1  
# ---------------------------------------------------------- 
host="localhost"
user="root" 
database="pr_laravel_8_vacancyonbert_2023_05"
mydb = mysql.connector.connect(host=host,user=user,password="",database=database) 
# ----------------------------------------------------------   
query = "SELECT "
query += " name "
query += " , count(*) " 
query += " FROM `dataset_jobtitle` "

query += " Where `perusahaan` = 'O' "
query += " and `jabatan` = 'O' "
query += " and `keterampilan_teknis` = 'O' "
query += " and `penempatan_wilayah` = 'O' "
query += " and `penempatan_provinsi` = 'O' "
query += " and `penempatan_kota` = 'O' "
query += " and `penempatan_kecamatan` = 'O' "
query += " and `penempatan_kelurahan` = 'O' "
query += " and `penempatan_kawasan` = 'O' "

query += " group by name "
query += " having count(*) = "+str(counter_row)+" "
query += " order by name asc"   
# query += " order by count(*) desc"  
# ----------------------------------------------------------     
print(query, flush=True)
# ----------------------------------------------------------   
mycursor = mydb.cursor()
mycursor.execute(query)
result =  mycursor.fetchall() 
# ---------------------------------------------------------- 
total_rows = len(result)
print("Total Row(s) : " + str(total_rows), flush=True) 
# ----------------------------------------------------------  
mycursor.close()
mydb.close()
# ---------------------------------------------------------- 
counter = 0 
# ----------------------------------------------------------
for x in result:    
    # ------------------------------------------------------
    counter         += 1  
    # ------------------------------------------------------ 
    nama        = str(x[0])     
    count       = str(x[1])      
    # ------------------------------------------------------  
    word = "[" + str(counter) + "/" +str(total_rows) + "] " 
    word += " __ " + nama           
    word += " __ " + count      
    print(word, flush=True)    
    # ------------------------------------------------------ 
# ----------------------------------------------------------
# ----------------------------------------------------------
# ----------------------------------------------------------
```
